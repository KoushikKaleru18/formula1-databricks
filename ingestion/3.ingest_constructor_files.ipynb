### Ingesting Constructors files

```python
constructor_schema = """
  constructorId INTEGER,
  constructorRef STRING,
  name STRING,
  nationality STRING,
  url STRING"""
```

```python
volume_dir = '/Volumes/formula1/default/f1_volume'
```

```python
constructor_df = spark.read \
.schema(constructor_schema) \
.json(f'{volume_dir}/raw/constructors.json')
```

```python
display(constructor_df)
```

```python
constructor_df.printSchema()
```

#### drop unwanted cols

```python
constructor_dropped_df = constructor_df.drop(constructor_df['url'])
```

#### Rename cols, and add new colm - ingestion_date

```python
from pyspark.sql.functions import current_timestamp
```

```python
constructor_final_df = constructor_dropped_df.withColumnRenamed("constructorId", "constructor_id") \
                                            .withColumnRenamed("constructorRef", "constructor_ref") \
                                            .withColumn("ingestion_date", current_timestamp())
```

```python
# display(constructor_final_df)
```

```python
constructor_final_df.write.mode('overwrite').parquet(f'{volume_dir}/processed/constructors')
```
